```python
import transformers
from transformers import AutoTokenizer
import torch
```

```python
model='meta-llama/Llama-2-7b-chat-hf'
```

```python
tokenizer=AutoTokenizer.from_pretrained(model, use_auth_token=True)
```

```python
from transformers import pipeline
```

```python
llama_pipeline=pipeline(
    'text-generation',    # LLAMA TASK # LLM TASK
    model=model,
    torch_dtype=torch.float16,
    device_map='auto'
)
```

```python
def get_llama_response(prompt:str)->None:
  sequences=llama_pipeline(prompt,
                           top_k=10,
                           eos_token_id=tokenizer.eos_token_id,
                           max_length=256
                           )
  print('chatbot:',sequences[0]['generated_text'])
```

```python
prompt1='what is the population of gaya district in india'
```

```python
get_llama_response(prompt1)
```

```python
get_llama_response('give me top 5 chicken food items which i can eat today')
```

```python
while True:
  user_input=input()
  if user_input.lower() in ['bye','quit','exit']:
    print('chatbot:Goodbye')
    break
  get_llama_response(user_input)
```
